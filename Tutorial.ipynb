```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
import pandas as pd
```

#### 데이터 : 페암 수술 환자 기록
* col : 종양의 유형, 폐활량, 호흡 곤란 여부, 고통 정도, 기침, 흡연, 천식 여부 등 수술 전 환자 상태
* Label : 수술 후 생존 결과 (1 생존 0 사망)

```python
df = pd.read_csv('thoracic_surgery.csv')
df
```

### Test Set, Training Set 나누기

```python
train,test = train_test_split(df,test_size = 0.2)
train = train.values
test = test.values
print(train.shape,test.shape)
```

```python
train_x,train_y = train[:,:17],train[:,17]
print(train_x.shape,train_y.shape)
```

```python
test_x,test_y = test[:,:17],test[:,17]
print(test_x.shape,test_y.shape)
```

### Model 설계 (간단한 모델)

```python
model = Sequential()
model.add(Dense(30,input_dim=17,activation='relu'))
model.add(Dense(1,input_dim=30,activation='sigmoid'))
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(train_x,train_y,epochs=30,batch_size=10)
```

```python
model.evaluate(test_x,test_y)
```
